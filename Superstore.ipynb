# Importing the Libraries

```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

# Importing the Data

```python
data=pd.read_csv('SampleSuperstore.csv')
data
```

# Checking for null data if any

```python
(data.isnull()).sum()
```

# Describing Data

```python
data.describe()
```

```python
data.nunique()
```

```python
sns.pairplot(data,hue="Ship Mode")
```

```python
sns.countplot(data=data,x="Ship Mode")
```

## Conclusion : Standard class is the most preferred Ship Mode

```python
plt.figure(figsize=[10,10])
plt.xticks(rotation=90)
sns.countplot(data=data,x="State")
```

## Conclusion : California is the state with highest number of consumers

```python
data['Category'].value_counts()
```

```python
plt.pie(data.Category.value_counts(),labels=data.Category.value_counts().index,textprops={'fontsize': 15}, startangle=90,
        autopct='%1.0f%%',colors=['#ff7675', '#ffeaa7', '#55efc4'],pctdistance=0.8,
        labeldistance=1.18)
centre_circle = plt.Circle((0, 0), radius=0.6, fc='white')
plt.gca().add_artist(centre_circle)
plt.show()
```

## Conclusion : Office Supplies are Sold approx 60%

```python
data['Sub-Category'].value_counts()
```

```python
plt.figure(figsize=[12,12])
plt.pie(data['Sub-Category'].value_counts(),labels=data['Sub-Category'].value_counts().index,textprops={'fontsize': 14}, startangle=90,
        autopct="%1.0f%%",
        pctdistance=0.8,
        labeldistance=1.18)
centre_circle = plt.Circle((0, 0), radius=0.6, fc='white')
plt.gca().add_artist(centre_circle)
plt.show()
```

```python
data['Quantity'].value_counts()
```

```python
sns.countplot(data.Quantity)
```

## Conclusion : Mostly 2-3 items are bought by consumers

```python
sns.pairplot(data,hue="Discount")
```

```python
data.Discount.value_counts()
```

```python
plt.figure(figsize=[10,10])
plt.pie(data['Discount'].value_counts(),textprops={'fontsize': 14}, startangle=90,
        pctdistance=0.8,
        labeldistance=1.18)
centre_circle = plt.Circle((0, 0), radius=0.6, fc='white')
plt.legend(labels=data['Discount'].value_counts().index)
plt.gca().add_artist(centre_circle)
plt.show()
```
